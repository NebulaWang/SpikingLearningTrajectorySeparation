```python
import os

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
```

```python
dat_dir = '/content/drive/MyDrive/Project/PlasticityDecoding/data'
SPK = []
ptype = []
for rnum in range(0, 30):
    SPK.append(np.load(dat_dir + f'/FixedRepresentation/Spk_an_wn_{rnum}.npy'))
    ptype.append(np.load(dat_dir + f'/FixedRepresentation/Ptype_an_wn_{rnum}.npy'))

SPK = np.concatenate(SPK, axis=0)
ptype = np.concatenate(ptype, axis=0)
```

```python
plt.figure(figsize=(2,1),dpi=600)
plt.hist(SPK[ptype == 0, 29 * 500 : 30 * 500].sum(axis=1) * 2, bins=30, range = (0, 30), density=True, color='deepskyblue', label='STDP')
# plt.xlabel('Firing rate (Hz)')
# plt.ylabel('Density')
# plt.legend()
plt.xticks([])
sns.despine()
plt.show()

plt.figure(figsize=(2,1),dpi=600)
plt.hist(SPK[ptype == 1, 29 * 500 : 30 * 500].sum(axis=1) * 2, bins=30, range = (0, 30), density=True, color='steelblue', label='BurstProp')
# plt.legend()
sns.despine()
plt.xticks([])
plt.show()

plt.figure(figsize=(2,1),dpi=600)
plt.hist(SPK[ptype == 2, 29 * 500 : 30 * 500].sum(axis=1) * 2, bins=30, range = (0, 30), density=True, color='green', label='Rein')
# plt.legend()
plt.ylabel('Density',fontsize=16)
sns.despine()
plt.xticks([])
plt.show()

plt.figure(figsize=(2,1),dpi=600)
plt.hist(SPK[ptype == 3, 29 * 500 : 30 * 500].sum(axis=1) * 2, bins=30, range = (0, 30), density=True, color='orange', label='BPTT')
plt.xticks([])
# plt.xlabel('Firing rate (Hz)')
# plt.ylabel('Density')
# plt.legend()
sns.despine()
plt.show()

plt.figure(figsize=(2,1),dpi=600)
plt.hist(SPK[ptype == 4, 29 * 500 : 30 * 500].sum(axis=1) * 2, bins=30, range = (0, 30), density=True, color='gray', label='NC')
# plt.legend()
plt.xlabel('Firing rate (Hz)',fontsize=16)
# plt.ylabel('Density',fontsize=16)
sns.despine()
plt.show()
```
