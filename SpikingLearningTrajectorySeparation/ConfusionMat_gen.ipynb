```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
import scipy as sp
import scipy.signal as signal
import torchvision.transforms.functional as tvF

from mpl_toolkits.axes_grid1 import make_axes_locatable

import math
```

```python
def confusion_draw_fixedrepresentation(confusion_dir, pic_dir):

    confusion_num = np.load(confusion_dir)
    confusion = np.round(confusion_num / confusion_num.sum(1), 3)

    sns.set_theme()

    # Draw a heatmap with the numeric values in each cell
    f, ax = plt.subplots(dpi=600,figsize=(6, 6))
    sns.heatmap(confusion, annot=True, annot_kws={'size':21}, linewidths=.5, ax=ax, cbar=False)
    plt.xticks([1 - 0.5, 2 - 0.5, 3 - 0.5, 4 - 0.5, 5 - 0.5], ['STDP', 'BDP', 'BPTT1', 'BPTT2', 'NC'], fontsize=21)
    plt.yticks([1 - 0.5, 2 - 0.5, 3 - 0.5, 4 - 0.5, 5 - 0.5], ['STDP', 'BDP', 'BPTT1', 'BPTT2', 'NC'], fontsize=21)

    plt.savefig(pic_dir)
    plt.show()

def confusion_draw_endtoend(confusion_dir, pic_dir):

    confusion_num = np.load(confusion_dir)
    confusion = np.round(confusion_num / confusion_num.sum(1), 3)

    sns.set_theme()

    # Draw a heatmap with the numeric values in each cell
    f, ax = plt.subplots(dpi=600,figsize=(6, 6))
    sns.heatmap(confusion, annot=True, annot_kws={'size':21}, linewidths=.5, ax=ax, cbar=False)
    plt.xticks([1 - 0.5, 2 - 0.5, 3 - 0.5, 4 - 0.5], ['STDP', 'BPTT1', 'BPTT2', 'NC'], fontsize=21)
    plt.yticks([1 - 0.5, 2 - 0.5, 3 - 0.5, 4 - 0.5], ['STDP', 'BPTT1', 'BPTT2', 'NC'], fontsize=21)

    plt.savefig(pic_dir)
    plt.show()

```

```python
save_dir = '/content/drive/MyDrive/Project/PlasticityDecoding/result'
for classifier in ['svm', 'TCNN', 'TTE']:
    for condition in ['fixedrepresentation', 'endtoend']:
        for test in ['in', 'out']:
            if test == 'in':
                confusion_dir = save_dir + f'/{classifier}_{condition}_confusion_num.npy'
            elif test == 'out':
                confusion_dir = save_dir + f'/{classifier}_{condition}_confusion_num1.npy'

            pic_dir = f'/content/drive/MyDrive/Project/PlasticityDecoding/confusion_pic/confusion_{condition}_{test}_{classifier}.png'

            if condition == 'fixedrepresentation':
                confusion_draw_fixedrepresentation(confusion_dir, pic_dir)
            elif condition == 'endtoend':
                confusion_draw_endtoend(confusion_dir, pic_dir)

            del confusion_dir, pic_dir
```
